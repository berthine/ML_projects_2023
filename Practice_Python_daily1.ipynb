Comment generer les données pour un model de classification

Supposons que tu as appris un nouveau model et que tu n'a pas des données réelles pour le tester. C'est possible de simuler les données avec Scikit-learn.

Les étapes à suivre sont les suivants:
1. Importer la librarie "from sklearn.datasets import make_classification"
2. Utiliser "make_classification()" pour créer lavarieté des données de classification. 
3. Il y a quelques possibilités telle que:

    * Generer les données avec une reponse binaire ou multiple classes.
    * Creér les reponses variables avec les classes balancées ou non. 
    * Produire des données qui sont difficile à classifier. 
    
Les parametres qu'il faut specifier  pour la fonction make_classification():

* n_samples: Specifier combien d'observations tu veux generer?
* n_features: Specifier les nombre des variables (attributs) numeriques que vous desirez. 
* n_classes: Le nombre de valuer unique pour la variable reponse.

Exemple: Prenons le cas où nous voulons generé les données avec 10000 observation dont la reponse variable est binare avec 6 attributs.

```python
X, y = make_classification(
    n_samples=10000, # 10000 observations 
    n_features=6, # 6 total attributs
    n_informative=4, # 4 attributs ou variables explicatives sont importantes
    n_classes=2, # la variable reponse est binaire 
    random_state=999 # la reproduction de ce données 
)
```

```python
##Nous pouvons transformer ces données en un tableau des données

# Créer le tableau des données avec les attributs comme colonnes
dataset = pd.DataFrame(X)
# Donner les nons aux variables
dataset.columns = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6']
# Ajouter la variable y comme colonne
dataset['y'] = y
#Affichier les informations du tableau crée
dataset.info()
```

```python
##fficher l'entete du tableau
dataset.head()
```

```python
#compter le nombre de valeur unique dans la variable y
dataset['y'].value_counts()
```

En voyant le nombres d'observation par classe, nous pouvons dire que les deux classes sont bien distribuées et il y a une balance entre les deux classes. 

Mai c'est possible de generer les données dont la variable reponse n'est pas balancée. Dans ce cas il suffit the specifier le pourcentage pour chaque classe. Voici une ilustration:

```python
attr, rep = make_classification(
    # les memes parametres, 1000 obsv, 6 attributs
    n_samples=1000, n_features=6, n_informative=4, n_classes=2, 
    # specifier le pourcentage de 0 et de 1 de la variable y. Par exemple  85% de 0 et 15% de 1
    weights=[0.85], 
)

donne = pd.DataFrame(attr)
# Donner les nons aux variables
donne.columns = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6']
# Ajouter la variable y comme colonne
donne['reponse'] = rep
donne.head()
```

```python
##verifions la distribution de 0 et 1
donne['reponse'].value_counts()
```

Nous pouvons voir que les deux classes ne sont pas equilibrées.
